```
import pandas as pd
mall = pd.read_csv("/content/Mall_Customers.csv")
```

```
import seaborn as sns
```

```
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
sns.countplot('Age',data=mall,hue='Genre')
plt.tight_layout()
```

```
sns.countplot('Genre',data=mall)
```

```
plt.figure(figsize=(12,6))
sns.countplot('Age',data=mall)
plt.tight_layout()
```

```
plt.figure(figsize=(20,6))
sns.countplot('Annual Income (k$)',data=mall,hue='Genre')
plt.tight_layout()
```

```
sns.distplot(mall['Annual Income (k$)'],bins=20)
```
