```
import pandas as pd
import matplotlib.pyplot as plt
```

```
weather=pd.read_csv('WeatherData.csv')
```

```
weather.head()
```

```
plt.plot(weather['Date'],weather['High Temperature'])
plt.xlabel('Days')
plt.ylabel('High Temperature')
plt.title('High Temperature vs Date',fontsize=20)
```

```
plt.plot(weather['Date'],weather['Low Temperature'])
plt.xlabel('Days')
plt.ylabel('Low Temperature')
plt.title('Low Temperature vs Date',fontsize=20)
```

```
plt.plot(weather['Date'],weather['Precipitation'])
plt.xlabel('Days')
plt.ylabel('Precipitation')
plt.title('Precipitation vs Date',fontsize=20)
```

```
plt.subplot(1,1,1)
plt.plot(weather['Date'],weather['High Temperature'])
plt.xlabel('Days')
plt.ylabel('High Temperature')
plt.title('High Temperature vs Date',fontsize=20)

plt.show()

plt.subplot(2,1,2)
plt.plot(weather['Date'],weather['Low Temperature'])
plt.xlabel('Days')
plt.ylabel('Low Temperature')
plt.title('Low Temperature vs Date',fontsize=20)

plt.show()
```
